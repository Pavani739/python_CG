```python
n = 10
present = 0
absent = 0
for i in range(1,n+1):
    val=int(input(f'enter Roll no {i} attendance:(0-absent,1-present)'))#formatting
    if(val==1):
        present+=1
    else:
        absent+=1
per = present/n*100
print("The presenties are",present)
print("The absenties are",absent)
print("The percentage of presenties",per)
```

```python
#index(sub_str): It returns the sub string index if it present in string otherwise it returns value error
s = "The sky is blue and Ocean is blue"
print(s.index("is"))
print(s.index("z"))#Here in this case it gives the value error
```

```python
#String Function
#strip(): It removes the leading(Front of the string) and lagging(Back of the string) white space
s="                The sky is blue and ocean is blue       "
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
#String Function
#'sep'.join(iterable): It joins the all Elements in the 'iterable' with separator
s= ['The', 'sky', 'is', 'blue', 'and', 'ocean', 'is', 'blue']
print(''.join(s))
print(' '.join(s))
print('@'.join(s))
```

```python
#String Checking Function
# where these checking Functions only gives the Output whether they are 'True/False'
str="Pavani123"
print(str.isupper())#It checks if the string is in captial
print(str.islower())#It checks if the string is in smaller
print(str.isdigit())#It checks if the string contains a digit or not 
print(str.isalpha())#It checks whether the string is alphabetic or not
print(str.isalnum())#It checks whether the string contains the digits , alphabetics
print(str.istitle())
print(str)
```

```python
#Check Strong Password
password = input("Enter a password:")
length = len(password)>=8
upper = False
lower = False
digit = False
special = False
for char in password:
    if char.isupper():
        upper=True
    elif char.islower():
        lower=True
    elif char.isdigit():
        digit=True
    else:
        special=True
if length and upper and lower and digit and special:
    print("Strong Password",password)
else:
    print("Weak Password",password)
```
