# Patterns

- Square pattern

```python
for i in range(5):
    for j in range(5):
        print("*",end=" ")
    print()
```

```python
for i in range(5):
    for j in range(5):
        print(i+1,end=" ")# here 'i' is the row
    print()
```

```python
for i in range(5):
    for j in range(5):
        print(j+1,end=" ")# here 'j is the coloumn 
    print()
```

- Lower RightAngle Triangle

```python
for i in range(5):
    for j in range(5):
        if i>=j:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if i>=j:
            print(j+1,end=" ")
        else:
            print(" ",end=" ")
    print()
```
